```python
#This code assigned zipcodes to crime records based on geolocation
from uszipcode import SearchEngine
search = SearchEngine(simple_zipcode=True)
crime_df['Zipcode'] = crime_df['Location'].apply(lambda x: search.by_coordinates(float(x.split(',')[0]), float(x.split(',')[1]), radius=5, returns=1)[0].zipcode)
crime_df.to_csv('../Chicago_Data/Chicago_Crimes_Geolocations_Rental_Prices', index=False)
```
